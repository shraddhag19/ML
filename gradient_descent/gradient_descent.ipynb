```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/advertising.csv')
data.head()
```

```python
plt.scatter(data['TV'], data['Sales'])
plt.xlabel('TV')

plt.ylabel('Sales')
plt.title('TV Advertising Budget vs Sales')
```

```python
X=data['TV']
Y=data['Sales']
```

```python
X=(X-X.mean())/X.std()
```

```python
m=0
b=0

L=0.01
epochs=1000
n=float(len(X))

for i in range(epochs):
    Y_pred=m*X+b
    Dm=(-2/n) * sum(X*(Y-Y_pred))
    Db=(-2/n) * sum(Y-Y_pred)
    m=m-L*Dm
    b=b-L*Db
print(m, b)
```

```python
plt.scatter(X,Y)
plt.plot(X, m*X+b, color='red')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('TV Advertising Budget vs Sales')
plt.show()
```
