```python
import pandas as pd
```

```python
data = pd.read_csv(r"advertising.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['TV'], data['Sales'])
plt.xlabel('TV')

plt.ylabel('Sales')
plt.title('TV Advertising Budget vs Sales')
```

```python
x=data['TV']
y=data['Sales']
```

```python
def calculateavg(x):
    return sum(x)/len(x)
```

```python
numsum=0
densum=0
avgvalx=calculateavg(x)
avgvaly=calculateavg(y)
for i in range(len(x)):
    temp1=x[i]-avgvalx
    temp2=y[i]-avgvaly
    numsum=numsum+temp1*temp2
    densum=densum+temp1*temp1
```

```python
numsum,densum
```

```python
m=numsum/densum
m
```

```python
b=avgvaly-m*avgvalx
b
```

```python
plt.scatter(x, y)
plt.plot(x, m*x+b, color='red')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('TV Advertising Budget vs Sales')
plt.show()
```
